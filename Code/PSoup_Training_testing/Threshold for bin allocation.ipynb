```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file (replace 'your_file.xlsx' with your actual file)
file_path = r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Training\Overall Training.xlsx'

sheet1 = pd.read_excel(file_path, sheet_name='Sheet1')
sheet2 = pd.read_excel(file_path, sheet_name='Sheet2')

# Clean column names for both sheets (remove leading/trailing spaces)
sheet1.columns = sheet1.columns.str.strip()
sheet2.columns = sheet2.columns.str.strip()

# First Graph: Bin Accuracy vs Model Name and Range to be Different
plt.figure(figsize=(10, 6))
for model_name in sheet1['Model Name'].unique():
    subset = sheet1[sheet1['Model Name'] == model_name]
    plt.bar(
        subset['Range to be different'].astype(str) + ' (' + model_name + ')',
        subset['Bin accuracy'],
        label=model_name
    )
plt.xlabel('Model Name and Range to be Different')
plt.ylabel('Bin Accuracy')
plt.title('Bin Accuracy vs Model Name and Range to be Different')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Legend on the right
plt.tight_layout()
plt.show()

# Second Graph: Kendall Similarities
plt.figure(figsize=(10, 6))
width = 0.2  # Bar width
x = range(len(sheet2['Model Name'].unique()))
models = sheet2['Model Name'].unique()

# Use the cleaned column names
good_sim = sheet2.groupby('Model Name')['Kendall good similarity'].mean()
moderate_sim = sheet2.groupby('Model Name')['Kendall moderate similarity'].mean()
similarity = sheet2.groupby('Model Name')['Kendall similarity'].mean()

plt.bar([i - width for i in x], good_sim, width=width, label='Kendall Good Similarity')
plt.bar(x, moderate_sim, width=width, label='Kendall Moderate Similarity')
plt.bar([i + width for i in x], similarity, width=width, label='Kendall Similarity')

plt.xticks(x, models, rotation=45, ha='right')
plt.xlabel('Model Name')
plt.ylabel('Similarity')
plt.title('Kendall Similarities')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Legend on the right
plt.tight_layout()
plt.show()

```

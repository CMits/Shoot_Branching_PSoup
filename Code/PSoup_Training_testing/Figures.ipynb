```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches

# Load the Excel file
excel_file_path = r'C:\Users\uqcmitsa\Desktop\Excell for final Paper\Updated_Automated_values_in_the_mutants.xlsx'
sheet_name = 'Sheet2'
df = pd.read_excel(excel_file_path, sheet_name=sheet_name)

# Extract the biological data, mutant names, simulation data, and actual values
biological_data = df['Biological'].values
mutant_names = df['Mutant'].values
simulation_data = df['Simulation'].values
actual_values = df['Actual_values'].values

# Ensure the data length matches the number of treatments
num_treatments = len(mutant_names)

# Define the range for "not significantly different" (within 10% tolerance)
lower_bound = 1 * 0.99
upper_bound = 1 * 1.01

#print(f"Range close to WT (within ±10%): {lower_bound:.2f} to {upper_bound:.2f}")

# Define the colors based on value thresholds and the defined range
def get_value_color(value, lower_bound, upper_bound):
    if value < lower_bound:
        return '#CC79A7'
    elif value > upper_bound:
        return '#009E73'
    else:
        return '#F0E442'

# Define the colors for biological and simulation data
def map_to_color(value):
    if value == -1:
        return '#CC79A7'
    elif value == 1:
        return '#009E73'
    else:
        return '#F0E442'

# Create the plot
fig, (ax_matrix, ax3) = plt.subplots(2, 1, figsize=(24, 15), gridspec_kw={'height_ratios': [2, 2]})

# Plot the biological significance matrix
for i, treatment in enumerate(mutant_names):
    ax_matrix.add_patch(plt.Rectangle((i - 0.5, 1), 1, 1, color=map_to_color(biological_data[i])))

# Plot the simulation significance matrix
for i, treatment in enumerate(mutant_names):
    bin_color = map_to_color(simulation_data[i])
    ax_matrix.add_patch(plt.Rectangle((i - 0.5, 0), 1, 1, color=bin_color))

# Add lines to distinguish each column and a horizontal line between the matrices
for i in range(num_treatments):
    ax_matrix.axvline(x=i - 0.5, color='black', linewidth=0.5)
ax_matrix.axhline(y=1, color='black', linewidth=1.5)

# Set the matrix plot properties
ax_matrix.set_xlim(-0.5, num_treatments - 0.5)
ax_matrix.set_ylim(0, 2)
ax_matrix.set_xticks([])  # Remove x-tick labels
ax_matrix.set_yticks([0.5, 1.5])
ax_matrix.set_yticklabels(['Simulation Bin', 'Biological Bin'], fontsize=30)


# Plot the data points for actual values as bins
bin_width = 0.4
for i, treatment in enumerate(mutant_names):
    bin_color = get_value_color(actual_values[i], lower_bound, upper_bound)
    ax3.add_patch(plt.Rectangle((i - bin_width/2, 0), bin_width, actual_values[i], color=bin_color))

# Add a horizontal line at y=1
ax3.axhline(y=1, color='black', linestyle='--')

# Plot the percentage tolerance lines
#ax3.axhline(y=lower_bound, color='red', linestyle=':')
#ax3.axhline(y=upper_bound, color='red', linestyle=':')

# Add labels and title
ax3.set_xlabel('Experimental Treatments', fontsize=35, labelpad=20)
ax3.set_ylabel('Simulation Actual Values', fontsize=35)
#ax3.set_title('Experimental Treatment Values', fontsize=20)

# Set x-axis limits to include all bins
ax3.set_xlim(-1, num_treatments)

# Set y-axis limits to 0-3
ax3.set_ylim(0, 2.5)

# Set y-axis tick label font size
ax3.tick_params(axis='y', labelsize=30)

# Set x-axis ticks to experimental treatments
ax3.set_xticks(range(0, num_treatments, 6))
ax3.set_xticklabels(mutant_names[::6], rotation=45, ha='right', fontsize=30)

# Remove the top and right spines
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)


# Save the bar plot
bar_plot_save_path = r'C:\Users\uqcmitsa\Desktop\Graphs\Validation_Graph.png'
plt.tight_layout()
plt.savefig(bar_plot_save_path, dpi=300)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Load data from Excel file
excel_file_path = r'C:\Users\uqcmitsa\Desktop\PSoup Analysis\Excell validation\DataBase.xlsx'
sheet_name = 'D53_N'

# Load the data from Excel
df = pd.read_excel(excel_file_path, sheet_name=sheet_name)

# Create a color map
def map_colors(value):
    if value == 1:
        return '#009E73'
    elif value == -1:
        return '#CC79A7'
    else:
        return '#F0E442'

# Apply color mapping to DataFrame
color_mapped_df = df.set_index('Mutant').applymap(map_colors)

# Define custom colormap
colors = ['#CC79A7', '#F0E442', '#009E73']
cmap = LinearSegmentedColormap.from_list('custom_map', colors, N=3)

# Transpose the DataFrame
color_mapped_df = color_mapped_df.transpose()

# Create a matrix plot
fig, ax = plt.subplots(figsize=(15, 5))  # Adjust the figure size
im = ax.imshow(color_mapped_df.replace({'#CC79A7': -1, '#F0E442': 0, '#009E73': 1}).values, aspect='auto', cmap=cmap, vmin=-1, vmax=1)

# Show ticks and labels with more space
plt.xticks(np.arange(len(color_mapped_df.columns)), color_mapped_df.columns, rotation=45, ha='right', fontsize=20)
plt.yticks(np.arange(len(color_mapped_df.index)), color_mapped_df.index, fontsize=12)

# Add grid lines to distinguish between rows
ax.set_xticks(np.arange(-0.5, len(color_mapped_df.columns), 1), minor=True)
ax.set_yticks(np.arange(-0.5, len(color_mapped_df.index), 1), minor=True)
ax.grid(which="minor", color="black", linestyle='-', linewidth=1)

# Add color bar
cbar = fig.colorbar(im, ax=ax)
cbar.set_ticks([-1, 0, 1])
cbar.set_ticklabels(['-1', '0', '1'])
cbar.set_label('Bin Value', fontsize=12)
cbar.ax.tick_params(labelsize=12)

# Calculate percentage of matching and non-matching bins
total_bins = color_mapped_df.size
matching_bins = np.count_nonzero(color_mapped_df.replace({'#CC79A7': -1, '#F0E442': 0, '#009E73': 1}).values == 1)
non_matching_bins = np.count_nonzero(color_mapped_df.replace({'#CC79A7': -1, '#F0E442': 0, '#009E73': 1}).values != 1)
matching_percentage = (matching_bins / total_bins) * 100
non_matching_percentage = (non_matching_bins / total_bins) * 100

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Define the file path for simulated data
simulated_file_path = r'C:\Users\uqcmitsa\Desktop\Excell for final Paper\Simulated_Ranking.xlsx'

# Read the simulated data
df_sim = pd.read_excel(simulated_file_path)

# Rank the simulated data based on Hormones.Sustained_growth
df_sim['Rank'] = df_sim['Hormones.Sustained_growth'].rank(method='first', ascending=True)

# Normalize the ranks to be in the range of 0 to 15
df_sim['Normalized Rank'] = (df_sim['Rank'] - df_sim['Rank'].min()) * (15 / (df_sim['Rank'].max() - df_sim['Rank'].min()))
df_sim['Normalized Rank'] = df_sim['Normalized Rank'].round().astype(int)

# Adjust the ranks so that the WT mutant is in the middle
wt_index = df_sim[df_sim['mutant'] == 'WT'].index[0]
middle_rank = 7  # Middle rank for 0 to 15 range
wt_rank = df_sim.loc[wt_index, 'Normalized Rank']
rank_shift = middle_rank - wt_rank
df_sim['Normalized Rank'] = df_sim['Normalized Rank'] + rank_shift

# Ensure the ranks are within the 0 to 15 range
df_sim['Normalized Rank'] = df_sim['Normalized Rank'].clip(lower=0, upper=15)

# Adjust the ranks based on base genotypes without "+"
for index, row in df_sim.iterrows():
    mutant = row['mutant']
    if '+' in mutant:
        base_mutant = mutant.split(',')[0].strip()
        base_rank = df_sim[df_sim['mutant'] == base_mutant]['Normalized Rank']
        if not base_rank.empty:
            df_sim.at[index, 'Normalized Rank'] = max(df_sim.at[index, 'Normalized Rank'], base_rank.iloc[0])

# Save the ranked simulated data to a new Excel file
output_simulated_file_path = r'C:\Users\uqcmitsa\Desktop\Excell for final Paper\Simulated_Ranking_Ranked.xlsx'
df_sim.to_excel(output_simulated_file_path, index=False)

print("Ranking complete. Ranked simulated data saved at:", output_simulated_file_path)

```

```python
import pandas as pd

# Load the CSV file (replace with your file path)
df = pd.read_csv(r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Testing\Benchmark_Model\Results_B_M_Test.csv')


# Create a new DataFrame to store the output
output_rows = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    psoup_treatment = row['Psoup treatment']
    compare_with = row['Compare with']
    
    # Get the sustained growth value for Psoup treatment
    psoup_value = row['Hormones.Sustained_growth']
    
    # Find the sustained growth value for the compare_with treatment
    compare_row = df[df['Psoup treatment'] == compare_with]
    
    if not compare_row.empty:
        compare_value = compare_row['Hormones.Sustained_growth'].values[0]
        
        # Compare the values and assign the bin
        difference = psoup_value - compare_value
        if difference > 0.05:
            bin_value = 1
        elif difference < -0.05:
            bin_value = -1
        else:
            bin_value = 0
    else:
        compare_value = None  # Set to None if compare_with treatment is not found
        bin_value = 0         # Default to 0 if compare_with treatment is not found
    
    # Add the result to the output list
    output_rows.append({
        'Psoup treatment': psoup_treatment,
        'Compare with': compare_with,
        'Hormones.Sustained_growth (Psoup)': psoup_value,
        'Hormones.Sustained_growth (Compare)': compare_value,
        'PSoup bin': bin_value
    })

# Convert the list to a DataFrame
output_df = pd.DataFrame(output_rows)

# Save the output to a new CSV file 
output_df.to_csv(r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Testing\Benchmark_Model\Comparison.csv', index=False)

print("Processing complete. Output saved to 'Comparison output.csv'.")
```

```python
import pandas as pd

# Load the CSV file (replace with your file path)
df = pd.read_csv(r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Testing\Benchmark_Model\Results_B_M_Test.csv')

# Create a new DataFrame to store the output
output_rows = []

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    psoup_treatment = row['Psoup treatment']
    compare_with = row['Compare with']
    hormone_to_compare = row['psoup_value']  # This column defines which hormone to compare

    # Skip rows where the value in 'psoup_value' is 'Skip'
    if hormone_to_compare == 'Skip':
        output_rows.append({
            'Psoup treatment': psoup_treatment,
            'Compare with': compare_with,
            'Hormone Compared': hormone_to_compare,
            f'{hormone_to_compare} (Psoup)': None,
            f'{hormone_to_compare} (Compare)': None,
            'PSoup bin': 0  # Assign bin value 0 for skipped rows
        })
        continue

    # Get the value of the hormone to compare for Psoup treatment
    psoup_value = row[hormone_to_compare] if hormone_to_compare in row else None

    # Find the corresponding row for the compare_with treatment
    compare_row = df[df['Psoup treatment'] == compare_with]

    if not compare_row.empty and hormone_to_compare in compare_row.columns:
        compare_value = compare_row.iloc[0][hormone_to_compare]

        # Compare the values and assign the bin
        difference = psoup_value - compare_value
        if difference > 0.05:
            bin_value = 1
        elif difference < -0.05:
            bin_value = -1
        else:
            bin_value = 0
    else:
        compare_value = None  # Set to None if compare_with treatment is not found
        bin_value = 0         # Default to 0 if compare_with treatment is not found

    # Add the result to the output list
    output_rows.append({
        'Psoup treatment': psoup_treatment,
        'Compare with': compare_with,
        'Hormone Compared': hormone_to_compare,
        f'{hormone_to_compare} (Psoup)': psoup_value,
        f'{hormone_to_compare} (Compare)': compare_value,
        'PSoup bin': bin_value
    })

# Convert the list to a DataFrame
output_df = pd.DataFrame(output_rows)

# Save the output to a new CSV file
output_df.to_csv(r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Testing\Benchmark_Model\Comparison.csv', index=False)

print("Processing complete. Output saved to 'Comparison.csv'.")
```

## Test different thresholds

```python
import pandas as pd

# Load the CSV file (replace with your file path)
df = pd.read_csv(r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Training\B_M_Twiked_mod\Results_B_M_Twiked_mod.csv')

# Define thresholds to test
thresholds = [0.01, 0.03, 0.05, 0.075, 0.1]
accuracy_results = []

# Iterate over each threshold
for threshold in thresholds:
    # Create a new DataFrame to store the output for this threshold
    output_rows = []

    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        psoup_treatment = row['Psoup treatment']
        compare_with = row['Compare with']
        hormone_to_compare = row['psoup_value']  # Column that tells which hormone to compare

        # Skip rows where the value in 'psoup_value' is 'Skip'
        if hormone_to_compare == 'Skip':
            output_rows.append({
                'Psoup treatment': psoup_treatment,
                'Compare with': compare_with,
                'Hormones.Sustained_growth (Psoup)': None,
                'Hormones.Sustained_growth (Compare)': None,
                'PSoup bin': 0  # Assign bin value 0 for skipped rows
            })
            continue

        # Get the sustained growth value for Psoup treatment
        psoup_value = row[hormone_to_compare]

        # Find the sustained growth value for the compare_with treatment
        compare_row = df[df['Psoup treatment'] == compare_with]

        if not compare_row.empty:
            compare_value = compare_row[hormone_to_compare].values[0]

            # Compare the values and assign the bin based on the current threshold
            difference = psoup_value - compare_value
            if difference > threshold:
                bin_value = 1
            elif difference < -threshold:
                bin_value = -1
            else:
                bin_value = 0
        else:
            compare_value = None  # Set to None if compare_with treatment is not found
            bin_value = 0         # Default to 0 if compare_with treatment is not found

        # Add the result to the output list
        output_rows.append({
            'Psoup treatment': psoup_treatment,
            'Compare with': compare_with,
            f'{hormone_to_compare} (Psoup)': psoup_value,
            f'{hormone_to_compare} (Compare)': compare_value,
            'PSoup bin': bin_value
        })

    # Convert the list to a DataFrame
    output_df = pd.DataFrame(output_rows)

    # Calculate accuracy (percentage of correct bins)
    correct_bins = output_df['PSoup bin'].value_counts().get(0, 0)
    total_bins = len(output_df)
    accuracy = correct_bins / total_bins * 100

    # Store the accuracy for this threshold
    accuracy_results.append({
        'Threshold': threshold,
        'Accuracy (%)': accuracy
    })

    # Save the output to a new CSV file for the current threshold
    output_filename = rf'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Training\B_M_Twiked_mod\Comparison_Threshold_{threshold}.csv'
    output_df.to_csv(output_filename, index=False)

# Convert accuracy results to a DataFrame and save
accuracy_df = pd.DataFrame(accuracy_results)
accuracy_df.to_csv(r'C:\Users\uqcmitsa\OneDrive - The University of Queensland\Desktop\Paper\Dataset\Training\B_M_Twiked_mod\Threshold_Accuracy.csv', index=False)

print("Processing complete. Accuracy results saved to 'Threshold_Accuracy.csv'.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches

# Load the Excel file
excel_file_path = #Add the file with the bin comparisons.
sheet_name = 'New'
df = pd.read_excel(excel_file_path, sheet_name=sheet_name)

# Extract the biological data, mutant names, simulation data, and actual values
biological_data = df['Biological'].values
mutant_names = df['Mutant'].values
simulation_data = df['Simulation'].values

# Ensure the data length matches the number of treatments
num_treatments = len(mutant_names)

# Define the range for "not significantly different" (within 10% tolerance)
lower_bound = 1 * 0.99
upper_bound = 1 * 1.01

# Define the colors based on value thresholds and the defined range
def get_value_color(value, lower_bound, upper_bound):
    if value < lower_bound:
        return '#CC79A7'
    elif value > upper_bound:
        return '#009E73'
    else:
        return '#F0E442'

# Define the colors for biological and simulation data
def map_to_color(value):
    if value == -1:
        return '#CC79A7'
    elif value == 1:
        return '#009E73'
    else:
        return '#F0E442'

# Create the plot
fig, ax_matrix = plt.subplots(1, 1, figsize=(24, 8))

# Plot the biological significance matrix
for i, treatment in enumerate(mutant_names):
    ax_matrix.add_patch(plt.Rectangle((i - 0.5, 1), 1, 1, color=map_to_color(biological_data[i])))

# Plot the simulation significance matrix
for i, treatment in enumerate(mutant_names):
    bin_color = map_to_color(simulation_data[i])
    ax_matrix.add_patch(plt.Rectangle((i - 0.5, 0), 1, 1, color=bin_color))

# Add lines to distinguish each column and a horizontal line between the matrices
for i in range(num_treatments):
    ax_matrix.axvline(x=i - 0.5, color='black', linewidth=0.5)
ax_matrix.axhline(y=1, color='black', linewidth=1.5)

# Set the matrix plot properties
ax_matrix.set_xlim(-0.5, num_treatments - 0.5)
ax_matrix.set_ylim(0, 2)
ax_matrix.set_xticks([])  # Remove x-tick labels
ax_matrix.set_yticks([0.5, 1.5])
ax_matrix.set_yticklabels(['Simulation Bin', 'Biological Bin'], fontsize=30)

# Add x-axis ticks for the mutant names every 5th mutant to reduce crowding
ax_matrix.set_xticks(range(0, num_treatments, 3))
ax_matrix.set_xticklabels(mutant_names[::3], rotation=45, ha='right', fontsize=20)

# Remove the top and right spines
ax_matrix.spines['right'].set_visible(False)
ax_matrix.spines['top'].set_visible(False)

# Save the matrix plot
matrix_plot_save_path = # add when you want to be saved
plt.tight_layout()
plt.savefig(matrix_plot_save_path, dpi=300)

plt.show()
```
